```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#file = pd.read_csv("226(normal).csv")
#file = pd.read_csv("1681815045818.csv")
file = pd.read_csv("1683794902904.csv")
file
```

```python
fig, ax = plt.subplots()
fig.set_figheight(15)
fig.set_figwidth(28)
for key1, grp1 in file.groupby(['numberRight', 'joinIteratorString']):
    #for key2, grp2 in grp1.groupby([]):
    ax = grp1.plot(ax=ax, kind='line', x='elementsRatio', y='execution time', label= str(key1))
plt.legend(loc='best')
plt.show()
```

```python
fig, ax = plt.subplots()
fig.set_figheight(15)
fig.set_figwidth(28)
for key1, grp1 in file.groupby(['numberRight', 'joinIteratorString']):
    #for key2, grp2 in grp1.groupby([]):
    if key1[0] != 500000 and key1[0] != 200000:
        ax = grp1.plot(ax=ax, kind='line', x='elementsRatio', y='execution time', label= str(key1))
plt.legend(loc='best')
plt.show()
```

```python
fig, ax = plt.subplots()
fig.set_figheight(15)
fig.set_figwidth(28)
for key1, grp1 in file.groupby(['ratio']):
    for key2, grp2 in grp1.groupby(['joinIteratorString']):
        ax = grp2.plot(ax=ax, kind='line', x='numberRight', y='execution time', label= str(key1) + "; " + str(key2))
plt.legend(loc='best')
plt.show()
#file.plot(x="numberRight", y="execution time")
```

```python
fig, ax = plt.subplots()
fig.set_figheight(15)
fig.set_figwidth(28)
for key1, grp1 in file.groupby(['ratio']):
    for key2, grp2 in grp1.groupby(['joinIteratorString']):
        ax = grp2.plot(ax=ax, kind='line', x='numberRight', y='rss total', label= str(key1) + "; " + str(key2))
plt.legend(loc='best')
plt.show()
#file.plot(x="numberRight", y="execution time")
```
